```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

folder1_path = '/scratch/ssd004/scratch/ziyuwang/ECCV2022_WaveGAN/cond'
folder2_path = '/scratch/ssd004/scratch/ziyuwang/ECCV2022_WaveGAN/results/animal_test/animal_test/fakes'

# Get image file names from folders
folder1_images = [f for f in os.listdir(folder1_path) if os.path.isfile(os.path.join(folder1_path, f))]
folder2_images = [f for f in os.listdir(folder2_path) if os.path.isfile(os.path.join(folder2_path, f))]

# Read the first 3 images from folder 1
img_list_1 = [mpimg.imread(os.path.join(folder1_path, img)) for img in folder1_images[:3]]

# Read the first 6 images from folder 2
img_list_2 = [mpimg.imread(os.path.join(folder2_path, img)) for img in folder2_images[:6]]

# Create a plot to display images
fig, ax = plt.subplots(1, 9, figsize=(20, 5))

# Display images from folder 1
for i, img in enumerate(img_list_1):
    ax[i].imshow(img)
    ax[i].axis('off')

# Display images from folder 2
for i, img in enumerate(img_list_2):
    ax[i+3].imshow(img)
    ax[i+3].axis('off')

plt.show()
```
